```python
import os
import numpy as np
import pandas as pd
from pathlib import Path
from IPython.display import display, HTML
display(HTML("<style>:root { --jp-notebook-max-width: 90% !important; }</style>"))


selection_csv = "/home/ubuntu/dev/teams-deepfake-detector-evaluation/assets/2025_07_16-selected_videos.csv"
recorded_videos_csv = "/data/video-streaming-windows/recorded-video-clips/2025_07_18-480x270_no_background_blur/2025_07_18-media_attributes.csv"
```

```python
selection_df = pd.read_csv(selection_csv)

# Reconstruct item_id and modifiers
# original_videos_df = pd.read_csv(original_videos_csv)
# original_videos_df["modifiers"] = original_videos_df["filepath"].map(lambda f: Path(f).parts[0])
# original_videos_df["item_id"] = original_videos_df["filepath"].map(lambda f: os.path.splitext(os.path.basename(f))[0])

recorded_videos_df = pd.read_csv(recorded_videos_csv)
recorded_videos_df[["item_id", "modifiers"]] = recorded_videos_df.apply(lambda row: os.path.splitext(os.path.basename(row["filepath"]))[0].split("_"), axis=1, result_type="expand")
```

```python
merged_df = selection_df[["item_id", "modifiers", "s3_object_key", "video_length", "duration", "frame_count"]].merge(
    right=recorded_videos_df[["item_id", "modifiers", "filepath", "duration", "frame_count"]],
    on=("item_id", "modifiers"),
    how="left",
    suffixes=("_original", "_recorded")
)

merged_df["duration_differs"] = np.abs(merged_df["duration_original"] - merged_df["duration_recorded"]) > 3

with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', None):
    display(merged_df)
```
